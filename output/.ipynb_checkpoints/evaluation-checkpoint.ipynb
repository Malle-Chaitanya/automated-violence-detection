```python
import numpy as np
import cv2
import os
from keras.models import load_model
from collections import deque

# Define your project directory
PROJECT_DIR = r"C:\Users\Vyshnavi Malle\Desktop\final_project"  # Update with your actual project path
MODEL_PATH = os.path.join(PROJECT_DIR, "modelnew.h5")  # Path to the model
VIDEO_DIR = os.path.join(PROJECT_DIR, "archive", "Real Life Violence Dataset", "Violence")  # Path to videos

def print_results(video, limit=None):
    # Ensure output directories exist
    output_dir = os.path.join(PROJECT_DIR, 'output')
    os.makedirs(output_dir, exist_ok=True)

    violence_frames_dir = os.path.join(output_dir, 'violence_frames')
    os.makedirs(violence_frames_dir, exist_ok=True)

    # Check if model exists
    if not os.path.exists(MODEL_PATH):
        print(f"❌ Model file NOT found at: {MODEL_PATH}")
        return

    print("✅ Loading model...")
    model = load_model(MODEL_PATH)  # Load the model
    Q = deque(maxlen=128)
    
    # Check if video file exists
    if not os.path.exists(video):
        print(f"❌ Video file NOT found at: {video}")
        return

    vs = cv2.VideoCapture(video)
    writer = None
    (W, H) = (None, None)

    total_frames = 0
    violence_frames = 0  # Counter for detected violence frames

    while True:
        (grabbed, frame) = vs.read()
        if not grabbed:
            break

        if W is None or H is None:
            (H, W) = frame.shape[:2]

        output = frame.copy()
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        frame = cv2.resize(frame, (128, 128)).astype("float32")
        frame = frame.reshape(128, 128, 3) / 255

        preds = model.predict(np.expand_dims(frame, axis=0))[0]
        Q.append(preds)
        results = np.array(Q).mean(axis=0)
        label = (preds > 0.50)[0]

        text_color = (0, 255, 0)  # Green (No Violence)

        if label:  # If violence detected
            text_color = (0, 0, 255)  # Red
            violence_frames += 1
            frame_name = os.path.join(violence_frames_dir, f"frame_{total_frames}.jpg")
            cv2.imwrite(frame_name, output)

        total_frames += 1

        text = "Violence: {}".format(label)
        FONT = cv2.FONT_HERSHEY_SIMPLEX
        cv2.putText(output, text, (5, 20), FONT, 0.75, text_color, 2)

        # Save processed video
        if writer is None:
            fourcc = cv2.VideoWriter_fourcc(*"MJPG")
            video_output_path = os.path.join(output_dir, "v_output.avi")
            writer = cv2.VideoWriter(video_output_path, fourcc, 30, (W, H), True)

        writer.write(output)

        # Display frame (only works in Jupyter Notebook)
        from IPython.display import display, Image
        _, buffer = cv2.imencode(".jpg", output)
        display(Image(data=buffer.tobytes()))

        key = cv2.waitKey(1) & 0xFF
        if key == ord("q"):
            break

    # Calculate percentage of violence detected
    violence_percentage = (violence_frames / total_frames) * 100 if total_frames > 0 else 0
    print(f"[INFO] Total Frames Processed: {total_frames}")
    print(f"[INFO] Violence Detected in {violence_frames} frames ({violence_percentage:.2f}%)")

    print("[INFO] Cleaning up...")
    if writer is not None:
        writer.release()
    vs.release()
    cv2.destroyAllWindows()

if __name__ == "__main__":
    video_path = os.path.join(VIDEO_DIR, "V_280.mp4")  # Update with your actual video file

    if os.path.exists(video_path):
        print_results(video_path)
    else:
        print(f"❌ Video file NOT found at: {video_path}")
```
